```julia
using Distributions,Plots,QuadGK
```

```julia
# lorenz curve
function lorenz_curve(d::UnivariateDistribution, t::Real)
    quadgk(x->quantile(d,x),0,t)[1]/quadgk(x->quantile(d,x),0,1)[1]
end

# Gini coefficient
function gini(d::UnivariateDistribution)
    1 - 2*quadgk(x->lorenz_curve(d,x),0,1)[1]
end
```

```julia

# 
d = Uniform(0, 1)

plot(t->lorenz_curve(d,t),0,1,fill=(0,0.5,:orange),label="gini = $(gini(d))",legend=:topleft)
plot!(t->t,label=false)
```
